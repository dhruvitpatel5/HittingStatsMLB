```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
file_path = 'baseball_hitting.csv'
df = pd.read_csv(file_path) #2508 entries 
df = df.dropna() # now, 2488 entries
df.head(355)
```

```python
target_players = ['B Harper', 'K Schwarber', 'N Castellanos']
player_stats = ['Runs', 'Hits', 'AVG', 'On-base Percentage', 'Slugging Percentage']
new_stats = df[player_stats]

# Get stats of target players 
for player in target_players:
    target_player_indexes = df[df['Player name'] == player].index
    target_player_stats = new_stats.loc[target_player_indexes].values.reshape(1, -1)
    
    # Calculate cosine similarity for each row in new_stats against target_player_stats
    def cos_sim(row, target_player_stats): 
        reshape = row.values.reshape(1, -1)
        return cosine_similarity(reshape , target_player_stats)
    new_stats['cosine_similarity'] = new_stats.apply(cos_sim,args=(target_player_stats,),axis=1)
    cos_stats = new_stats.sort_values(by='cosine_similarity', ascending=False).head(10)

    # Print results
    print('Players similar to ', player, ":")
    for index in cos_stats.index:
        similarity_score = new_stats.loc[index, 'cosine_similarity']
        print(df.loc[index, 'Player name'], similarity_score) 
```
